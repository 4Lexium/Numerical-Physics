# Maxwell- hastighetsfordeling

```python
#Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import statistics 
from matplotlib.ticker import PercentFormatter
import matplotlib.ticker as ticker
%matplotlib inline
```

```python
#extracting 2D list with t, x, y vals from given file

def read_file(filename):
    file = open(filename, "r")
    lines = file.readlines()
    
    file.seek(0)
    file.truncate
    file.seek(0) 
    
    file_vls = []
    for line in file:
        vls = line.split()
        file_vls.append(vls)
        
    file.close()
    file_vls = file_vls[2:]
    return file_vls
```

```python
velocity_array = []
V_X_list = []
V_Y_list = []
V_ABS_list = []
X_list = []
Y_list = []

for i in range(1, 50):
    filename = "mass8_" + str(i) + ".txt"
    file_vals = read_file(filename)
    velocity_values = []
    for m in range(1, int(len(file_vals))):
        x_m = float(file_vals[m][1])*0.22
        x_mp = float(file_vals[m-1][1])*0.22
        y_m = float(file_vals[m][2])*0.22
        y_mp = float(file_vals[m-1][2])*0.22
        t_m = float(file_vals[m][0])
        t_mp = float(file_vals[m-1][0])
        v_x_value = (x_m-x_mp)/(t_m-t_mp)
        v_y_value = (y_m-y_mp)/(t_m-t_mp)
        V_value = np.sqrt(v_x_value**2+v_y_value**2)

        X_list.append(x_m)
        Y_list.append(y_m)
        V_X_list.append(v_x_value)
        V_Y_list.append(v_y_value)
        V_ABS_list.append(V_value)
        velocity_values.append([v_x_value, v_y_value, V_value])
    velocity_array.append(velocity_values)

"""
veloctity_array includes subarrays for data on velocity in a 3D array
velocity_array demiensions = #(3velocty vars) * #(number of measurements in given file) * #(49 files)
additional subarrays [lists] seperate componental velocities from absulute speed and positions [cm] and [s]
V_X_lsit, V_Y_list, V_ABS_lsit, X_list and Y_list each with 920 elements

"""
print(len(V_X_list))
print(len(V_Y_list))
print(len(V_ABS_list))

print(len(X_list))
print(len(Y_list))
```

# Numerisk del

```python
#setting up histograms and bar/scatterplots
N = 920
M = int(round(N**(1/3)*2, 0)+1)  #customary bins using the Rice method 

#Converting lists into numpy arrays
V_X_array = np.array(V_X_list)
V_Y_array = np.array(V_Y_list)
V_ABS_array = np.array(V_ABS_list)

X_array = np.array(X_list)
Y_array = np.array(Y_list)
```

```python
#visualisation Histograms
plt.style.use("fivethirtyeight")
plt.rcParams['figure.dpi'] = 100
fig1, (ax2) = plt.subplots(nrows = 1, ncols = 1)
fig2, (ax4) = plt.subplots(nrows = 1, ncols = 1)
fig3, (ax6) = plt.subplots(nrows = 1, ncols = 1)

ax2.hist(V_ABS_array, bins = M, color = "black", edgecolor="green")
ax2.yaxis.set_major_formatter(ticker.PercentFormatter(xmax=len(V_ABS_array)))
ax2.set_title("Distribution for Absolute Velocity")
ax2.set_xlabel("|V| [cm/s]")
ax2.set_ylabel("frequency [%]")

ax4.hist(V_X_array, bins = M, color = "black", edgecolor="green")
ax4.yaxis.set_major_formatter(ticker.PercentFormatter(xmax=len(V_X_array)))
ax4.set_title("Distribution for Velocity x component")
ax4.set_xlabel("V_x [cm/s]")
ax4.set_ylabel("frequency [%]")

ax6.hist(V_Y_array, bins = M,color = "black", edgecolor="green")
ax6.yaxis.set_major_formatter(ticker.PercentFormatter(xmax=len(V_Y_array)))
ax6.set_title("Distribution for Velocity y component")
ax6.set_xlabel("V_y [cm/s]")
ax6.set_ylabel("frequency [%]")

fig1.tight_layout()
fig2.tight_layout()
fig3.tight_layout()
plt.show()
```

```python
#visualisation Scatter and Barplots
plt.rcParams['figure.dpi'] = 100
fig4, (ax1) = plt.subplots(nrows = 1, ncols = 1)
fig5, (ax2) = plt.subplots(nrows = 1, ncols = 1)

ax1.scatter(V_X_array, V_Y_array, color = "black")
ax1.set_title("Distribution for Velocity x and y components")
ax1.set_xlabel("V_x [cm/s]")
ax1.set_ylabel("V_y [cm/s]")

ax2.plot(X_array, Y_array, "x", color = "green")
ax2.set_title("Distribution for x and y coordinates")
ax2.set_xlabel("X [cm]")
ax2.set_ylabel("Y [cm]")

fig4.tight_layout()
fig5.tight_layout()
plt.show()
```

# Analytisk del

```python
def velocity_abs_distribution(v_list, dv):
    v_min = min(v_list)
    v_max = max(v_list)
    v_mean = statistics.mean(list((np.array(v_list))**2))
    B = 1 / v_mean
    v_plot_x = np.arange(v_min, v_max, dv)
    v_plot_y = 2*B*v_plot_x*np.exp(-B*v_plot_x**2)/np.pi                     
    return v_plot_x, v_plot_y

def velocity_component_distribution(v_list, dv):
    v_min = min(v_list)
    v_max = max(v_list)
    v_mean = statistics.mean(list((np.array(v_list))**2))
    B = 1 / v_mean
    v_plot_x = np.arange(v_min, v_max, dv)
    v_plot_y = ((B/np.pi)**2) * np.exp(-B*v_plot_x**2)                
    return v_plot_x, v_plot_y
```

```python
def normalize_v(v_list, dv):
    Sum_v = 0
    for i in v_list:
        Sum_v += abs(i)*dv
    a = 1 / Sum_v
    return list(np.array(v_list)*a)
```

```python
V_abs_x, V_abs_y = velocity_abs_distribution(V_ABS_list, 0.5)
V_abs_y = normalize_v(V_abs_y, 0.5)
V_compx_x, V_compx_y = velocity_component_distribution(V_X_list, 0.5)
V_compx_y = normalize_v(V_compx_y, 0.5)
V_compy_x, V_compy_y = velocity_component_distribution(V_Y_list, 0.5)
V_compy_y = normalize_v(V_compy_y, 0.5)

fig6, (ax1) = plt.subplots(nrows = 1, ncols = 1)
fig7, (ax2) = plt.subplots(nrows = 1, ncols = 1)
fig8, (ax3) = plt.subplots(nrows = 1, ncols = 1)

ax1.plot(V_abs_x, V_abs_y, color = "green")
ax1.set_title("Theoretical Distribution for Absolute Velocity")
ax1.set_xlabel("|V| [cm/s]")
ax1.set_ylabel("frequency")

ax2.plot(V_compx_x, V_compx_y, color = "green")
ax2.set_title("Theoretical Distribution for Velocity x component")
ax2.set_xlabel("V_x [cm/s]")
ax2.set_ylabel("frequency")

ax3.plot(V_compy_x, V_compy_y, color = "green")
ax3.set_title("Theoretical Distribution for Velocity y component")
ax3.set_xlabel("V_y [cm/s]")
ax3.set_ylabel("frequency")

fig6.tight_layout()
fig7.tight_layout()
fig8.tight_layout()
plt.show()
```

```python
v_abs_mean_rms = np.sqrt(statistics.mean(list(np.array(V_ABS_list)**2)))
v_abs_mean = statistics.mean(V_ABS_list)

v_x_mean_rms = np.sqrt(statistics.mean(list(np.array(V_X_list)**2)))
v_x_mean = statistics.mean(V_X_list)

v_y_mean_rms = np.sqrt(statistics.mean(list(np.array(V_Y_list)**2)))
v_y_mean = statistics.mean(V_Y_list)

print(f"sqrt()/2          = {np.sqrt(np.pi) / 2}")
print(f"E(V_abs)/V_abs_rms = {v_abs_mean / v_abs_mean_rms}")
```
